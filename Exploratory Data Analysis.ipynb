## To check Null Values

```python
plt.figure(figsize=(12,4))
sns.heatmap(data.isnull(),cbar=False,yticklabels=False,cmap='summer')
plt.show()
```

# Checking Correlation

```python
plt.figure(figsize=(15,8))
sns.heatmap(data.corr(),annot=True,cmap='viridis')
plt.show()
```

### Key Observations:
    Among all the feature "Pagevalues" have high correlation with the Target variable.
    Most of the columns have positive correlation with the target variable.Only few have negative correlation.
    Among that "ExitRates" have high correlation.
    Most of the features have multicollinearity.Top 3 very strongly correlated Features are
        ('BounceRates' & 'ExitRates' is strongly correlated = 0.91)
        ('ProductRelated' & 'ProductRelated_Duration' is strongly correlated = 0.86)
        ('Administrative' & 'Administrative_Duration' is strongly correlated = 0.61)

# Checking Outliers

```python
Numerical_List =['Administrative', 'Administrative_Duration', 'Informational','Informational_Duration',
                 'ProductRelated', 'ProductRelated_Duration', 'BounceRates', 'ExitRates',
                 'PageValues', 'SpecialDay','OperatingSystems', 'Browser', 'Region', 'TrafficType']
```

```python
plt.figure(figsize=(14,15))
for i in range(0,len(Numerical_List)):
    plt.subplot(4,4,i+1)
    sns.boxplot(data[Numerical_List[i]],color='green',orient='v')
    plt.title(Numerical_List[i])
```

    All the features in the dataset have outliers

# Checking  Skewness

```python
plt.figure(figsize=(28,20))
for i in range(0,len(Numerical_List)):
    plt.subplot(4,4,i+1)
    sns.distplot(data[Numerical_List[i]],color='green',kde=True) 
    plt.title(Numerical_List[i])
```

     Like we infered in the summary statistics all of the features are Right or positively skewed.
     There is no normal distribution in the data

## Bivariate analysis of each categorical column

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Month',data=data,hue='Revenue')
plt.show()
```

    From the above plot we can infer lot of browsing time has happened for the months May,November,December,March
    For Feburary there were no Revenue generated but browsing has happened.
    The overall conversion rate of browsing time to purchase is really low for all the months.

```python
plt.figure(figsize=(8,4))
sns.countplot(x='Weekend',data=data,hue='Revenue')
plt.show()
```

     Lot of browsing time is on week days and the conversion rate is also higher when compared to weekends.

```python
plt.figure(figsize=(8,4))
sns.countplot(x='VisitorType',data=data,hue='Revenue')
plt.show()
```

     The Returning Visitors have browsed more than a new visitor. 
     The purchase rate of the returning visitor is really low when compared with the new visitors

# Bivariate Analysis

```python
plt.figure(figsize=(14,15))
for i in range(0,len(Numerical_List)):
    plt.subplot(4,4,i+1)
    sns.boxplot(data['Revenue'],data[Numerical_List[i]])
    plt.xlabel(None)
    plt.title(Numerical_List[i])
```

```python
sns.pairplot(data[['Administrative', 'Administrative_Duration', 'Informational','Informational_Duration',
                 'ProductRelated', 'ProductRelated_Duration', 'BounceRates', 'ExitRates',
                 'PageValues', 'SpecialDay','OperatingSystems', 'Browser', 'Region', 'TrafficType']])
plt.show()
```

# Outliers In Each Column

```python
print('Length of each column which has values above 0.99 quantile')
for i in range(len(Numerical_List)):
    Leng = len(data[data[Numerical_List[i]] > data[Numerical_List[i]].quantile(0.99)])
    if Leng > 0:
        print(Numerical_List[i],' = ',Leng)
```
